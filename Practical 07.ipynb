```python
text ="Tokenization is the first step in text analytics.The process of breaking down a text paragraph into smallerchunks such as words or sentences is called Tokenization."
```

```python
from nltk.tokenize import sent_tokenize
tokenized_text = sent_tokenize(text)
print("Sentence Tokenization:", tokenized_text)
```

```python
from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(text)
print("Word Tokenization:", tokenized_word)
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
print("Stop Words:", stop_words)
```

```python
text = "How to remove stop words with NLTK library in Python?"
text = re.sub('[^a-zA-Z]', ' ', text)
```

```python
tokens = word_tokenize(text.lower())  
print("Tokenized Sentence:", tokens)
```

```python
filtered_text = [w for w in tokens if w not in stop_words]
print("Filtered Sentence:", filtered_text)
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
e_words = ["wait", "waiting", "waited", "waits"]
```

```python
print("Stemming Results:")
for w in e_words:
    rootWord = ps.stem(w)
    print(w, "->", rootWord)
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
text = "studies studying cries cry"
tokenization = nltk.word_tokenize(text)
```

```python
print("Lemmatization Results:")
for w in tokenization:
    print("Lemma for {} is {}".format(w, wordnet_lemmatizer.lemmatize(w)))
```
