```python
import matplotlib.pyplot as plt
```

```python
pt=[20,25,28,30,80,40,45,50,60,25]
pt1=[30,15,80,48,50,60,70,39,80,99]
plt.plot(pt)
plt.plot(pt1)
plt.show()
```

```python
plt.plot(pt,c="r",ls="-.",mec="b",mfc="y",marker="X")
plt.plot(pt1,c="y",ls="-",mec="y",mfc="g",marker="p")
plt.xlabel("Sample Points")
plt.ylabel("values")
plt.title("Values vs Points")
plt.legend(["A", "B"])
plt.show()
```
